# download pose net weights

```python
import requests
import xml.etree.cElementTree as ET
import os
```

```python
tf_models_url = "https://storage.googleapis.com/tfjs-models/"
proxies = {
    "http": "agent.baidu.com:8118",
    "https": "agent.baidu.com:8118"
}
r = requests.get(tf_models_url, proxies=proxies)
res_xml = r.text
```

```python
tree = ET.fromstring(res_xml)
tree
```

```python
weights_uris = list()
for child in tree:
    if child.tag == "{http://doc.s3.amazonaws.com/2006-03-01}Contents":
        if child[0].text.find("weights/posenet/") > -1 :
            weights_uris.append(child[0].text)
```

```python
weights_uris
```

```python
weights_urls = map(lambda x: tf_models_url + x, weights_uris)
```

```python
weights_urls
```

```python
local_dir = "/Users/baidu/Documents/git/tfjs-models"
```

```python
download_status = set()
```

```python
for i, weights_uri in enumerate(weights_uris):
    if weights_uri in download_status:
#         print("INFO: skip {}/{} {}".format(i+1, len(weights_uris), weights_uri))
        continue
    save_path = os.path.join(local_dir, weights_uri)
    save_dir = os.path.dirname(save_path)
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    url = tf_models_url + weights_uri
    proxies = {
        "http": "agent.baidu.com:8118",
        "https": "agent.baidu.com:8118"
    }
    print("INFO: download {}/{} {}".format(i+1, len(weights_uris), weights_uri))
    try:
        r = requests.get(url, proxies=proxies)
        if r.status_code == 200:
            with open(save_path, "wb") as fp:
                fp.write(r.content)
                download_status.add(weights_uri)
        else:
            print("ERROR: download {}/{} {}".format(i+1, len(weights_uris), weights_uri))
    except Exception as e:
        print("EXCEPTION: download {}/{} {} {}".format(i+1, len(weights_uris), weights_uri, e.message))
```

```python
for i, weights_uri in enumerate(weights_uris):
    if weights_uri in download_status:
#         print("INFO: skip {}/{} {}".format(i+1, len(weights_uris), weights_uri))
        continue
    print("INFO: download {}/{} {}".format(i+1, len(weights_uris), weights_uri))
```
